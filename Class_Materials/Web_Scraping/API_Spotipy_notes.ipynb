```python
# import config let you load the 'config.py' file that is in the same folder of the notebook and contains your id and password

import config
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials


sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=config.client_id, client_secret=config.client_secret))

results = sp.search(q ="Lose yourself",limit=50,market="GB")

results
```

```python
# to access the link with id of a song
### you enter in the dictionaries (nested -> one in another one) you go with brackets

song_link = results["tracks"]["items"][0]["external_urls"]["spotify"]
song_link

# you put [0] when you have a list in the dictionary.
```

```python
song_id = results["tracks"]["items"][0]["id"]
song_id
```
